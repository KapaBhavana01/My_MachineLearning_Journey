Simple Linear Regression Model Implemtation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# creating small dataset
x_train = np.array([1.0,2.0,3.0])
y_train = np.array([200,300,400])
m = x_train.shape[0]
```

```python
#visualize the data
plt.scatter(x_train,y_train, marker="x", c="g", label="data")
plt.title("Visualize the data")
plt.xlabel("size")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
#model building and plotting to check whether the straight line fits good or not
w = 100 
b=100
def calculatefun(x,w,b):
    m=x.shape[0]
    f=np.zeros(m)
    for i in range(m):
        f[i]=x[i]*w+b
    return f
#plotting f 
f_val = calculatefun(x_train,w,b,)
plt.plot(x_train,f_val, c="r", label ="function line")
plt.scatter(x_train,y_train, marker="x", c="g", label="data")
plt.title("Visualize the data")
plt.xlabel("size")
plt.ylabel("price")
plt.legend()
plt.show()
```

```python
#predicting y value using linear regression formula
w=100
b=100
x_train=2.0
y_pred=w*x_train+b
print("Predicted value")
print(y_pred)
```
